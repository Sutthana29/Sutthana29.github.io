#### Plot Creation using Python, Altair and vega-lite for HW5

##### how to include plots from Vega-lite editor


```python
import pandas as pd
import numpy as np
import altair as alt
alt.data_transformers.disable_max_rows()
```

```python
license_url = 'https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/licenses_fall2022.csv'
```

```python

license = pd.read_csv(license_url, parse_dates = ["Effective Date"])
license.head()
```

```python
# add a brush
brush = alt.selection_interval(encodings=['x', 'y'])

chart1 = alt.Chart(license_url).transform_calculate(
    EffectiveYear="year(datum['Effective Date'])"
).transform_bin(
    'YearBin', 'EffectiveYear',bin=alt.Bin(step=10) 
).mark_rect().encode(
    alt.X('YearBin:O', title="Effective Year (binned)"),
    alt.Y('State:O'),
    alt.Color("count():Q",scale=alt.Scale(scheme='blues'))
).properties(
    height=400
).add_params(
    brush
)


#chart1

chart2 = alt.Chart(license_url).transform_calculate(
    EffectiveYear="year(datum['Effective Date'])"
).transform_bin(
    'YearBin', 'EffectiveYear'
    ).mark_bar().encode(
    alt.X('License Type:O'),
    alt.Y('count():Q')
).transform_filter(
    brush
)


chartURL = (chart1.properties(width=300)) | (chart2.properties(width=300))

chartURL
```
